```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
Container_LEDon=pd.read_csv("DarkBox/TestContainer_LEDon.txt",sep=' ')
Container_LEDoff=pd.read_csv("DarkBox/TestContainer_LEDoff.txt",sep=' ')
LAB_LEDon=pd.read_csv("DarkBox/TestLAB.txt",sep=' ')
Water_LEDon=pd.read_csv("DarkBox/TestWater.txt",sep=' ')
QuantumDot_LEDon=pd.read_csv("DarkBox/TestQuantumDot.txt",sep=' ')
pedestaldatas=pd.read_csv("DarkBox/TestPedestal.txt",sep=' ')
channel_consideration="chz"
bins_num="auto"
#datas=datas[channel_consideration]
#legend_lines = [f"{stat}: {value:.2f}" for stat, value in desc.items()]
#for i in legend_lines:
#    plt.plot("",label=i)
plt.hist(Container_LEDon[channel_consideration], bins=bins_num,histtype='step',label="LED on"+channel_consideration)
plt.hist(Container_LEDoff[channel_consideration], bins=bins_num,histtype='step',label="LED off"+channel_consideration)
plt.hist(LAB_LEDon[channel_consideration], bins=bins_num,histtype='step',label="LAB"+channel_consideration)
plt.hist(Water_LEDon[channel_consideration], bins=bins_num,histtype='step',label="Water"+channel_consideration)
plt.hist(QuantumDot_LEDon[channel_consideration], bins=bins_num,histtype='step',label="QuantumDot"+channel_consideration)
plt.hist(pedestaldatas[channel_consideration], bins=bins_num,histtype='step',label="Pedestal"+channel_consideration)

plt.title("Dark box setting")
plt.xlabel("QDC counts")
plt.ylabel("Counts")
plt.xlim(0,700)
plt.ylim(100,200)
plt.legend()
plt.show()
```
